```r
df <- read.csv("hv_rankings_results.csv",header=TRUE,sep=',')
# df <- subset(df,!is.na(rank))
# df <- subset(df,dataset!='505_tecator')
summary(df,maxsum=18)
```

```r
res <- pairwise.wilcox.test(df$rank_hv.auditor_fp_violation.ave_precision_score, 
                            df$method_nice, 
                            p.adjust.method = 'bonferroni',
                            paired = T)
cap = paste0("Bonferroni-adjusted $p$-values using a Wilcoxon signed rank test of  (FP-Violation, APS) hypervolume scores ",
      "for the methods across all problems. Bold: $p<$0.05.")
library(xtable)
res$p.value.formatted = format(res$p.value,scientific=T,digits=2)
res$p.value.formatted = ifelse(res$p.value < 0.05, paste0("{\\bf ",res$p.value.formatted, "}"), 
                                  res$p.value.formatted)
# print(res$p.value.formatted)
ltx <- xtable(res$p.value.formatted, caption=cap, display=c(rep('s',8)),
              caption.placement="top",sanitize.text.function = identity)


print(ltx,sanitize.text.function = identity) # enchancement: print to file, input to latex
#print file
filename ='rank_hv.auditor_fp_violation.ave_precision_score.csv' 
# cat(paste0(rownames(res$p.value), collapse = ','), file = filename, append = T, sep = '\n')
cat(paste0(colnames(res$p.value), collapse = ','), file = filename, append = T, sep = '\n')
cat(paste0(rownames(res$p.value), ',', apply(res$p.value,1,paste0, collapse=','),collapse='\n'), 
    file = filename, append = T, sep = '\n')
#display nice table
ltx
```
